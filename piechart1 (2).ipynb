```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16, 6), subplot_kw=dict(aspect="equal"))


data = [923280,629095,315341,55255,16306,7857]
ingredients = ["Europe","North America","Asia","South America","Africa","Oceania"]


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

ingredients1 = ["Europe: 47.4%","North America: 32.3%","Asia :16.2%","South America: 2.8%","Africa: 0.8%","Oceania: 0.4%"]
wedges = ax.pie(data)

ax.legend(ingredients1,title="Continents")

#plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Comparison of Total cases")

plt.show()
```
