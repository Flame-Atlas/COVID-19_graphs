```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = ('COVID19', 'EBOLA', 'MERS', 'SARS')
y_pos = np.arange(len(objects))
performance = [1923848,28646,2494,8096]
plt.bar(y_pos, performance, align='center', alpha=1,edgecolor='black')
plt.xticks(y_pos, objects)
plt.xlabel('Epidemic names --- >')
plt.ylabel('Number of Cases --->')
plt.title('Comparison of Total cases of various Epidemics')

plt.show()
```
